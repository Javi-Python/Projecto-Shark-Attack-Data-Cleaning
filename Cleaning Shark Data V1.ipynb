# Lets Start by Cleaning the Activity Column

```python
sharks_h1_drop.Activity.unique()
```

There are way to many not fitting values to our variable, we are going to try and extract as much information from the non standarized columns as we can. Lets start by creating a working copy of our subseted data frame to work on

```python
sharks_copy = sharks_h1_drop.copy()
```

From SRC we import the funciones file which contains all our data cleaning functions

```python
import SRC.funciones as fu
```

We create a list of regex terms and a list of what we want to change those regex terms too

```python
list_regex = [r".*?\bswimming\b.*",r'.*?\bsurfing\b.*', r'.*?\bpaddling\b.*', r'.*?\bfishing\b.*', r'.*?\bspearfishing\b.*', r'.*?\bwading\b.*', r'.*?\bscuba\b\s+\bdiving\b.*', r'.*?\bsnorkeling\b.*', r'.*?\bwalking\b.*', r'.*?\bfree\b\s+\bdiving\b.*', r'.*?\bbody\b\s+\bboarding\b.*', r'.*?\bstanding\b.*', r'.*?\bbathing\b.*']
to_change = ['swimming', 'surfing', 'paddling', 'fishing', 'spearfishing', 'wading', 'scuba diving', 'snorkeling', 'walking', 'free diving', 'body boarding', 'standing', 'bathing']
```

We apply our replace_regex function to the activities column in order to extract the pertinent data from it and create a new column with said values

```python
sharks3 = fu.replace_regex(sharks_copy, list_regex, to_change, 'Activity_Regex', 'Activity' )
```

As we can see the value counts for the top activities have increased after our data cleaning

```python
sharks3.Activity_Regex.value_counts().head(10)
```

# Lets Clean the Age Column

In order to clean the Age Column we will use a different function of the funciones.py to extract only the numbers from the age column and create a new column with them. We will start by creating a regex pattern that should encompass all numbers

```python
number_regex = r'([0-9]+)'
```

```python
sharks_age = fu.extract_regex(sharks3, 'Age_Regex', 'Age', number_regex)
```

```python
sharks_age.head()
```

```python
sharks_age.Age_Regex.value_counts().head(10)
```

We turn the values in the age column to numeric values

```python
pd.to_numeric(sharks_age['Age_Regex'])
```

# Lets Clean The Sex Column

As we previously did for activities, we will clean the sex column using the replace_regex function. We start by creating two lists


```python
regex_sex = [r"m ", r".*?\blli\b.*"]
sex = ['m', 'NaN']
```

We apply our function

```python
sharks_replace_sex = fu.replace_regex(sharks_age, regex_sex, sex, 'Sex_Regex', 'Sex')
```

```python
sharks_replace_sex.nunique()
```

As we can see the Sex column is left with only 3 unique values, one of which is a NAN

# Lets Clean the Species Column


To clean the Species column we will be using our replace_regex function to extract the values that fit the species that we will be focusing on and pass them to a new column

```python
species_regex = [r'.*?\bwhite\b\s+\bshark\b.*',r'.*?\bblue\b\s+\bshark\b.*', r'.*?\btiger\b\s.*',
           r'.*?\bbull\b\s.*',r'.*?\bshark\b\s+\binvolvement\b.*',r'.*?\bwobbegong\b\s+\bshark\b.*',
           r'.*?\bblacktip\b\s.*', r'.*?\bbronze\b\s+\bwhaler\b.*', r'.*?\bmako\b\s.*',r'.*?\bnurse\b\s.*',
           r'.*?\bhammerhead\b\s.*', r'.*?\braggedtooth\b\s.*']
species_change = ['White shark', 'Blue shark', 'Tiger shark', 
         'Bull shark', 'Not a shark', 'Wobbegong shark',
         'Blacktip shark','Bronze whaler shark', 'Mako shark', 
         'Nurse shark', 'Hammerhead shark', 'Raggedtooth shark']
```

```python
Clean_Sharks = fu.replace_regex(sharks_replace_sex, species_regex, species_change, 'Species_Regex', 'Species')
```

```python
Clean_Sharks.head(10)
```

```python
Clean_Sharks2 = fu.replace_regex(Clean_Sharks, fatal_regex, fatal_change, 'Fatal_Regex', 'Fatal_Y_N')
```
